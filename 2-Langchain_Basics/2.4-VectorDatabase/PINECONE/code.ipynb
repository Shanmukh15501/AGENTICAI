```python
from pinecone import Pinecone
import os
```

```python
pc = Pinecone(
    api_key=os.getenv("PINECONE_API_KEY"),
)
```

```python
from pinecone import ServerlessSpec

# Managed by cloud provider
if not pc.has_index("my-index1"):
    pc.create_index(
        name="my-index1",
        dimension=1024,
        metric="cosine",
        spec=ServerlessSpec
        (
            cloud="aws",
            region="us-east-1"
        )
    )
```
